```python
from matplotlib import pyplot as plt
import random
import numpy as np

def sumOfFiveHundred():
    ans=0
    times=500
    while(times > 0):
        ans = ans + random.uniform(0, 1)
        times = times-1
    return ans

times = 50000
templist =[]
while(times > 0):
    templist.append(sumOfFiveHundred())
    times = times-1
    
#X = np.array(templist)

plt.hist(templist,density=True, bins=500, alpha=0.5)
plt.show()
print("The shape of the histogram turned out to be pretty similar to a normal density function.")


        
```

```python
from scipy.stats import uniform
from scipy.stats import norm
import matplotlib.pyplot as plt
from math import sqrt
import numpy as np

b = sqrt(3)
loc   = -1 * b
scale =  2 * b

x=np.linspace(-4,4,200000)
y = uniform.pdf(x,loc,scale)
y1 = norm.pdf(x,0,1)


plt.plot(x,y,label='uniform',alpha=1)
plt.plot(x,y1,label='normal',alpha=1)
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
from scipy.stats import expon
from scipy.stats import uniform
import numpy as np


def exponentialGraph(n,labda,bins):
    labda=(1/labda)
    U=uniform.rvs(size=n)
    X=expon.ppf(U,loc=0,scale=labda)
    X=np.sort(X)
    plt.hist([X],bins,density=True,label="inverseCDF",alpha=0.8)
    plt.plot(X, expon.pdf(X,0,labda), label="Actual")
    plt.legend()
    plt.show()
    return X

exponentialGraph(n=10000,labda=1.5, bins=300)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import norm
import numpy as np


def normalGraph(n,mean,stdeviation,bins):
    U=uniform.rvs(size=n)
    X=norm.ppf(U,loc=0,scale=stdeviation)
    X=np.sort(X)
    plt.hist(X,density=True, bins=bins, alpha=0.8, label="inverseCDF")
    plt.plot(X, norm.pdf(X,mean,stdeviation), alpha=1, label="Actual")
    
    plt.legend()
    plt.show()
    return X

normalGraph(n=10000,mean=0,stdeviation=3.0, bins=200)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import rayleigh
import numpy as np


def rayleighGraph(n=1,mean=1,stdeviation=1, bins = 50):
    U=uniform.rvs(size=n)
    X=rayleigh.ppf(U,loc=mean,scale=stdeviation)
    X=np.sort(X)
    plt.hist(X,density=True, bins=bins, alpha=0.8, label="inverseCDF")
    plt.plot(X, rayleigh.pdf(X,mean,stdeviation), label="Actual")
    plt.legend()
    plt.show()
    return X

rayleighGraph(n=10000,mean=0,stdeviation=1.0, bins=250)
```
